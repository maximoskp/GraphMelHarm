```python
import pandas as pd
import os
```

```python
os.makedirs('figs', exist_ok=True)
```

```python
files_to_compare = [
    'results/FiLM/graph.csv',
    'results/HyperNetwork/graph.csv',
]
```

```python
import matplotlib.pyplot as plt
import re

figs_dir = 'figs'
os.makedirs(figs_dir, exist_ok=True)

dfs = []
labels = []
for path in files_to_compare:
    df = pd.read_csv(path)
    dfs.append(df)
    labels.append(path.split('/')[1])

common_columns = dfs[0].columns.tolist()[2:]
for column in common_columns:
    plt.figure(figsize=(8, 5))
    for df, label in zip(dfs, labels):
        plt.plot(df[column].values, marker='o', label=label)

    plt.title(column)
    plt.xlabel('Index')
    plt.ylabel(column)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()

    safe_name = re.sub(r'[^0-9A-Za-z_.-]+', '_', column)
    output_path = os.path.join(figs_dir, f'{safe_name}.png')
    plt.savefig(output_path, dpi=300)
    plt.close()
```
